## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
#Identifying the outliers
df.boxplot('Rating')
````

````python
#Removing the outliers in ratings field.
df['Rating'] = df['Rating'].clip(lower=0, upper=5)
#print(df['Rating'].describe())


#Replacing the null values in 'Rating' field with 'mean'
mean_rating = df['Rating'].mean()
df['Rating'].fillna(mean_rating, inplace=True)

# Replacing null values in 'Type' column with 'Unknown'
df['Type'].fillna('Unknown', inplace=True)

# Replacing null values in 'Content Rating' column with 'Not Specified'
df['Content Rating'].fillna('Not Specified', inplace=True)

# Replacing null values in 'Current Version' column with 'Unknown'
df['Current Ver'].fillna('Unknown', inplace=True)

# Replacing null values in 'Android Version' column with 'Unknown'
df['Android Ver'].fillna('Unknown', inplace=True)

# Missing Values/Null Values Count after data manipulation
print(df.isnull().sum())
````

````python
#Converting 'Price' into a numeric
df['Price'] = df['Price'].astype(str).str.replace('$', '').astype(float)
df.to_csv('/content/drive/MyDrive/Projects/EDA/Play_Store_Data.csv', index=False)
````

````python
#Converting 'Reviews' into a numeric value
df['Reviews'] = df['Reviews'].astype(str).str.replace('M', '').astype(float)
df['Reviews'] = df['Reviews'].astype(float).astype(int)
df.to_csv('/content/drive/MyDrive/Projects/EDA/Play_Store_Data.csv', index=False)
````

````python
#Converting 'Installs' into a numeric value
df['Installs'] = df['Installs'].astype(str).str.replace(',', '')  # Remove commas
df['Installs'] = df['Installs'].str.replace('+', '')  # Remove plus signs
df['Installs'] = df['Installs'].str.replace(' ','')  # Remove spaces
df['Installs'] = df['Installs'].replace('Free', '0')  # Replace 'Free' with '0'
df['Installs'] = df['Installs'].astype(float)  # Convert to float
df.to_csv('/content/drive/MyDrive/Projects/EDA/Play_Store_Data.csv', index=False)

````

````python
index_to_delete = 10474
df = df.drop(index_to_delete)
df.to_csv('/content/drive/MyDrive/Projects/EDA/Play_Store_Data.csv', index=False)
````

````python
df['Last Updated'] = df['Last Updated'].astype(str)
df.to_csv('/content/drive/MyDrive/Projects/EDA/Play_Store_Data.csv', index=False)
````

````python
#Creating a new 'User Sentiment' field
def get_sentiment(rating):
    if rating >= 4.0:
        return 'Positive'
    elif rating == 3.0:
        return 'Neutral'
    else:
        return 'Negative'

df['Sentiment'] = df['Rating'].apply(get_sentiment)
df.to_csv('/content/drive/MyDrive/Projects/EDA/Play_Store_Data.csv', index=False)
````

````python
# Chart - 11 visualization code
# Calculating sentiment distribution by category
sentiment_distribution = df.groupby(['Category', 'Sentiment']).size().unstack().fillna(0)
sentiment_distribution = sentiment_distribution.div(sentiment_distribution.sum(axis=1), axis=0)

# Stacked bar plot
sentiment_distribution.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Sentiment Distribution by App Category')
plt.xlabel('App Category')
plt.ylabel('Proportion of Sentiment')
plt.legend(title='Sentiment')
plt.xticks(rotation=90)
plt.show()

````
